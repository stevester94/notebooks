```python
%matplotlib widget

import matplotlib
import math
from matplotlib import pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import HTML
```

```python
duration_secs = 1
sample_rate = 100
f = 2




# PLOT
fig, (message_ax, carrier_ax, carrier_slice_ax) = plt.subplots(3, figsize=(9,9))

# Tuples are (<start>, <stop>, <steap>)
@widgets.interact(phase_0=(0,2*np.pi,0.01), phase_1=(0,2*np.pi,0.01))
def interact(phase_0, phase_1):
    message_ax.clear()

    T = np.linspace(0, duration_secs, duration_secs*sample_rate)
    x_0 = np.sin(2*np.pi*f*T + phase_0)
    x_1 = np.sin(2*np.pi*f*T + phase_1)
    # Plot message signal
    message_ax.plot(T,x_0, color="r", label="Signal 0")
    message_ax.plot(T,x_1, color="b", label="Signal 1")

    message_ax.set_title("Message Signal")
    message_ax.set_xlabel("Time")
    message_ax.set_ylabel("Amplitude")
    message_ax.grid()
    message_ax.legend()
```
