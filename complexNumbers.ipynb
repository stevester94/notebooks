```python
import matplotlib.pyplot as plt
import numpy as np

fs = 100
duration_secs = 2
t = np.linspace(0,duration_secs, int(duration_secs*fs))

in_A      = 2
in_f_Hz   = 10
in_phi    = 0
in_sig    = in_A * np.exp( 2j * np.pi * in_f_Hz * t + in_phi)

# Create a Figure
fig = plt.figure()
fig.set_figheight(20)
fig.set_figwidth(20)

# Set up Axes
ax1 = fig.add_subplot(411)
# ax1.plot(cos_out, label='PLL Output')
ax1.plot(in_sig.real, label='Input Signal (real)')
ax1.plot(in_sig.imag, label='Input Signal (imag)')

"""
OK This conceptually makes sense; the magnitude of the phasor is constant.
"""
ax1.plot(np.abs(in_sig), label='Input Signal (mag)')
ax1.set_title('Waveforms')

plt.grid()
plt.legend()
```
