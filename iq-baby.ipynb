So we have a weird signal being generated by A(t) and phi(t).

We know the functions for both amplitude and phase of the original signal. Using this we can easily plug them into the formulas for I(t) and Q(t)

```python
import numpy as np
import matplotlib.pyplot as plt

fig, (graph_1, graph_2) = plt.subplots(2,1)

fig.set_figheight(7)
fig.set_figwidth(25)

# Linearly increasing amplitude
def amplitude(t):
    amplitude_freq_Hz=1
    return np.sin(2*np.pi*t*amplitude_freq_Hz)

# OK SO WE KNOW WHAT A(t) IS AND PHI(t) IS, WE CAN REVERSE THIS INTO I(t) AND Q(t)
def phase(t):
    phase_freq_Hz=10
    return np.sin(2*np.pi*t*phase_freq_Hz)

def I(t):
    phase_freq_Hz=10
    return amplitude(t)*np.cos(phase(t))


def Q(t):
    phase_freq_Hz=10
    return amplitude(t)*np.sin(phase(t))

carrier_freq_Hz=10
duration_secs=3

time = np.arange(0, duration_secs, 0.001) 
sine_signal_1   = [amplitude(t) * np.sin(2 * np.pi * carrier_freq_Hz * t + phase(t)) for t in time]
quadrature_signal = [I(t)*np.sin(2 * np.pi * carrier_freq_Hz * t) + Q(t)*np.sin(2 * np.pi * carrier_freq_Hz * t + (np.pi/2)) for t in time]

graph_1.plot(time, sine_signal_1)
graph_2.plot(time, quadrature_signal)
plt.show()
```
