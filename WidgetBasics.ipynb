```python
%matplotlib widget
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML
```

```python
rng = np.random.default_rng(1337)
def generate_noise(noise_mean, noise_stddev, n=1000):
    """
    loc == mean,
    scale == stddev
    """
    noise = rng.normal(loc=noise_mean, scale=noise_stddev, size=n)
    
    
    return noise

def generate_sine_and_time(num_periods=2, num_points=-1, f=1, amp=1, phi=0):
    """
    
    """
    w = 2*np.pi*f
    if num_points == -1:
        num_points = int(np.ceil(w*2))
    x = np.linspace(0,(1/f)*num_periods, num_points)
    y = amp*np.sin(w * (x-phi))
    return x,y
```

# Basics
**All axes need to have unique names for the charts to operate independently**

```python
fig,ax = plt.subplots()
# Note how ipywidgets adds its figure name
fig.suptitle("My Figure")

# Tuples are (<start>, <stop>, <steap>)
@widgets.interact(noise_mean=(0, 10, 1), noise_stddev=(0, 4, .1))
def noise(noise_mean = 0, noise_stddev=1.0):
    """Remove old lines from plot and plot new one"""
    ax.clear()
    
    noise = generate_noise(noise_mean, noise_stddev)
    ax.scatter(noise, range(len(noise)), s=2)
```

## Multiple charts and sliders
