Numpy operates in terms of radians.
Recall that Pi=180 degrees

$$ sin(0) = 0 $$
$$ sin(\pi / 2) = 1 $$
$$ sin(\pi) = 0 $$
$$ sin(3 \pi / 4) = -1 $$
$$ sin(2\pi) = 0 $$

2 Pi is one full cycle

$$ \textrm{Equation for sine wave} $$
$$ A * \sin(2\pi * f * t + \phi) $$
$$ f\textrm{: Frequency, oscillations per second}$$
$$ A\textrm{: Amplitude} $$
$$ t\textrm{: time} $$
$$ \phi\textrm{: Phase offset} $$

```python
import numpy as np
import matplotlib.pyplot as plt

duration_secs = 2
frequency     = 2
phase         = 0
amplitude     = 1

t = np.arange(0, duration_secs, 0.01) 
sine_signal   = amplitude * np.sin(2 * np.pi * frequency * t + phase)

plt.plot(t, sine_signal)
plt.show()
```

Cosine is just sine phase shifted -90 degrees (Half pi)

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 3 * np.pi, 0.1) # create x array of angels from range 0 to 3*3.14
sine_signal   = np.sin(t)
cosine_signal = np.cos(t)


plt.plot(t, sine_signal)
plt.plot(t, cosine_signal)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 3 * np.pi, 0.1) # create x array of angels from range 0 to 3*3.14
sine_signal   = np.sin(t)
cosine_signal = np.cos(t - np.pi/2)


plt.plot(t, sine_signal)
plt.plot(t, cosine_signal)
plt.show()
```

# Alright its complex time
Complex sinusoids can be represented as complex exponentials
$$x = Ae^{2 \pi f t j + j\phi}$$

Note that the real component, the in-phase component, is a cosine

Instantaneous phase is the angle between the phasor and the x axis. I don't think it denotes actual phase (Since you'd have to compare it with another signal... right?

```python
sample_rate = 100
duration_secs = 2
f = 1.0
phi = 0
A = 2.0

T = np.linspace(0,duration_secs, sample_rate*duration_secs)
x = A * np.e ** (2.0j * np.pi * f * T + phi)

# PLOT
fig, (ax, magnitude_ax, phase_ax, polar_ax) = plt.subplots(4, figsize=(9,15))

# Plot signal
ax.plot(T,x.real, color="r", label="Signal.real")
ax.plot(T,x.imag, color="b", label="Signal.imag")

ax.set_title("Complex Signal")
ax.set_xlabel("Time")
ax.set_ylabel("Amplitude")
ax.grid()
ax.legend()

# Plot magnitude
magnitude_ax.plot(T,np.abs(x), color="r", label="|Signal|")

magnitude_ax.set_title("Complex Signal Magnitude")
magnitude_ax.set_xlabel("Time")
magnitude_ax.set_ylabel("Magnitude")
magnitude_ax.grid()
magnitude_ax.legend()

# Plot phase
phase_ax.plot(T,np.angle(x), color="r", label="arg(Signal)")

phase_ax.set_title("Complex Signal Instantaneous Phase")
phase_ax.set_xlabel("Time")
phase_ax.set_ylabel("Phase(Radians)")
phase_ax.grid()
phase_ax.legend()

# Plot polar
polar_ax.plot(x.real,x.imag, color="r", label="Signal")

polar_ax.set_title("Complex Signal Polar Coordinates")
polar_ax.set_xlabel("Real")
polar_ax.set_ylabel("Imag")
polar_ax.grid()
polar_ax.legend()

fig.tight_layout()
```
